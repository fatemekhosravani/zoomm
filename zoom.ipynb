```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = plt.imread("img.jpg")

plt.imshow(img)
plt.show()
```

```python
def zoom_image(img, scale_factor):

    h, w = img.shape[:2]
    
    h2 = int(np.floor(h * scale_factor))
    w2 = int(np.floor(w * scale_factor))
    
    y_coords = (np.arange(h2) / scale_factor).astype(int)
    x_coords = (np.arange(w2) / scale_factor).astype(int)

    y= np.clip(y_coords, 0, h - 1)
    x = np.clip(x_coords, 0, w - 1)

    yy, xx = np.meshgrid(y, x, indexing='ij')

    zoomed = img[yy, xx, :]

        
    return zoomed
```

```python
plt.figure(figsize=(15, 5))


plt.subplot(131)
z05 = zoom_image(img, 0.5)
plt.imshow(z05)
plt.title('Scale: 0.5x')
plt.axis('on')


plt.subplot(132)
z1 = zoom_image(img, 1.0)
plt.imshow(z1)
plt.title('Scale: 1.0x (Original)')
plt.axis('on')


plt.subplot(133)
z15 = zoom_image(img, 1.5)
plt.imshow(z15)
plt.title('Scale: 1.5x')
plt.axis('on')

plt.tight_layout()
plt.show()
```
